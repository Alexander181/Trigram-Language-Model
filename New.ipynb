Unigram Code Into Trigram

```python
# Import necessary libraries
import nltk
import random
from nltk import trigrams
from collections import Counter, defaultdict
import pandas as pd

nltk.download("punkt")
nltk.download("gutenberg")

# Load and preprocess corpus (first 1,000 sentences of Austen's "Emma")
from nltk.corpus import gutenberg
corpus = gutenberg.sents("austen-emma.txt")[:1000]
flat_corpus = [word.lower() for sentence in corpus for word in sentence if word.isalpha()]

# Trigram Model
trigram_counts = Counter(trigrams(flat_corpus))
total_trigrams = sum(trigram_counts.values())

# Trigram Prediction Function
def predict_trigram(sequence):
    # Predict next word based on the last two words' trigram probabilities
    last_bigram = tuple(sequence[-2:])  # Last two words in the sequence
    possible_trigrams = {trigram[2]: count for trigram, count in trigram_counts.items() if trigram[:2] == last_bigram}
    if possible_trigrams:
        next_word = random.choices(list(possible_trigrams.keys()), weights=possible_trigrams.values())[0]
    else:
        next_word = random.choice(list(trigram_counts.keys()))[0]  # fallback to a random word if no trigram available
    return next_word

# Sample sequences to test the Trigram Model
sample_sequences = [
    ["she", "was"],
    ["he", "had"],
    ["they", "were"],
    ["it", "is"],
    ["i", "think"],
    ["you", "should"],
    ["we", "are"],
    ["this", "was"],
    ["do", "you"],
    ["how", "can"]
]

# Predict next words based on trigram model
results = []
for sequence in sample_sequences:
    trigram_prediction = predict_trigram(sequence)
    results.append({
        "Sequence": " ".join(sequence),
        "Trigram": trigram_prediction
    })

# Convert results to DataFrame for display
results_df = pd.DataFrame(results)
results_df
```
